```python
import numpy as np
import torch
import matplotlib.pyplot as plt
np.set_printoptions(suppress=True)
```

```python
def compute_pos_encoding(seq_length, output_dim, n = 10000):
    pos_matrix = np.zeros((seq_length, output_dim))
    for seq_index in range(0, seq_length):
        for index in range(0, output_dim//2):
            x = seq_index/(n**((2*index)/output_dim))
            (pos_matrix[seq_index, 2*index],
             pos_matrix[seq_index, 2*index + 1]) = np.sin(x), np.cos(x)
    return pos_matrix
```

```python
sequence_length = 6
output_dimension = 4
position_matrix = compute_pos_encoding(sequence_length, output_dimension)
```

```python
print(position_matrix)
```

```python
img_ax = plt.matshow(position_matrix)
plt.gcf().colorbar(img_ax)
plt.show()
```

```python
pos_mat = compute_pos_encoding(50, 256)
cax = plt.matshow(pos_mat)
plt.gcf().colorbar(cax)
plt.show()
```
